# Cricket Player Performance Analysis

**Goal:** Analyze player and team performance to identify match-winning factors

**Tools:** Python (Pandas, NumPy), Matplotlib, Seaborn

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("Libraries imported successfully!")
```

## 2. Load Data

```python
# Load cricket data
df = pd.read_csv(r'C:\Users\lenovo\OneDrive\Desktop\Projects\cricket-player-performance\cricket_data.csv')

print(f"Data loaded successfully!\nShape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
```

## 3. Exploratory Data Analysis

```python
# Display first few rows
df.head(10)
```

```python
# Data information
df.info()
```

```python
# Statistical summary
df.describe()
```

```python
# Check for missing values
print("Missing Values:")
print(df.isnull().sum())
```

## 4. Batting Performance Analysis

```python
# Top run scorers
top_scorers = df.groupby('Player_Name')['Runs'].sum().sort_values(ascending=False).head(10)
print("Top 10 Run Scorers:")
print(top_scorers)

# Visualization
plt.figure(figsize=(12, 6))
top_scorers.plot(kind='barh', color='skyblue')
plt.xlabel('Total Runs', fontsize=12)
plt.ylabel('Player', fontsize=12)
plt.title('Top 10 Run Scorers', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Best batting averages (minimum 500 runs)
batsmen = df[df['Runs'] > 500].groupby('Player_Name').agg({
    'Runs': 'sum',
    'Average': 'mean',
    'Strike_Rate': 'mean',
    'Centuries': 'sum',
    'Half_Centuries': 'sum'
}).sort_values('Average', ascending=False).head(10)

print("Top 10 Batsmen by Average:")
batsmen
```

```python
# Strike Rate vs Average scatter plot
batsmen_data = df[df['Runs'] > 500]

plt.figure(figsize=(10, 6))
scatter = plt.scatter(batsmen_data['Average'], batsmen_data['Strike_Rate'], 
                     c=batsmen_data['Centuries'], s=100, cmap='viridis', alpha=0.6)
plt.xlabel('Batting Average', fontsize=12)
plt.ylabel('Strike Rate', fontsize=12)
plt.title('Batting Average vs Strike Rate', fontsize=14, fontweight='bold')
plt.colorbar(scatter, label='Centuries')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 5. Bowling Performance Analysis

```python
# Top wicket takers
bowlers = df[df['Wickets'] > 0]
top_wickets = bowlers.groupby('Player_Name')['Wickets'].sum().sort_values(ascending=False).head(10)
print("Top 10 Wicket Takers:")
print(top_wickets)

# Visualization
plt.figure(figsize=(12, 6))
top_wickets.plot(kind='barh', color='lightcoral')
plt.xlabel('Total Wickets', fontsize=12)
plt.ylabel('Player', fontsize=12)
plt.title('Top 10 Wicket Takers', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Best bowlers by average (minimum 30 wickets)
quality_bowlers = bowlers.groupby('Player_Name').agg({
    'Wickets': 'sum',
    'Bowling_Average': 'mean',
    'Economy': 'mean'
})
quality_bowlers = quality_bowlers[quality_bowlers['Wickets'] > 30].sort_values('Bowling_Average').head(10)

print("Best Bowlers by Average:")
quality_bowlers
```

## 6. Team Performance Analysis

```python
# Team-wise total runs
team_stats = df.groupby('Team')['Runs'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
team_stats.plot(kind='bar', color='teal', alpha=0.7)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Total Runs', fontsize=12)
plt.title('Team-wise Total Runs', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

```python
# Win/Loss analysis by team
team_performance = df.groupby(['Team', 'Match_Result']).size().unstack(fill_value=0)
team_performance['Win_Percentage'] = (team_performance['Won'] / 
                                      (team_performance['Won'] + team_performance['Lost']) * 100).round(2)

print("Team Win/Loss Record:")
team_performance.sort_values('Win_Percentage', ascending=False)
```

## 7. Match-Winning Factors Analysis

```python
# Compare performance in wins vs losses
won_matches = df[df['Match_Result'] == 'Won']
lost_matches = df[df['Match_Result'] == 'Lost']

print("Average Performance in Winning vs Losing Matches:")
comparison = df.groupby('Match_Result').agg({
    'Runs': 'mean',
    'Average': 'mean',
    'Strike_Rate': 'mean',
    'Wickets': 'mean',
    'Economy': 'mean'
}).round(2)

comparison
```

```python
# Visualization: Won vs Lost comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Average Runs
win_loss_runs = df.groupby('Match_Result')['Runs'].mean()
win_loss_runs.plot(kind='bar', ax=axes[0], color=['green', 'red'], alpha=0.7)
axes[0].set_xlabel('Match Result', fontsize=12)
axes[0].set_ylabel('Average Runs', fontsize=12)
axes[0].set_title('Average Runs: Won vs Lost', fontsize=14, fontweight='bold')
axes[0].tick_params(axis='x', rotation=0)
axes[0].grid(True, alpha=0.3, axis='y')

# Average Strike Rate
win_loss_sr = df.groupby('Match_Result')['Strike_Rate'].mean()
win_loss_sr.plot(kind='bar', ax=axes[1], color=['green', 'red'], alpha=0.7)
axes[1].set_xlabel('Match Result', fontsize=12)
axes[1].set_ylabel('Average Strike Rate', fontsize=12)
axes[1].set_title('Average Strike Rate: Won vs Lost', fontsize=14, fontweight='bold')
axes[1].tick_params(axis='x', rotation=0)
axes[1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

## 8. Correlation Analysis

```python
# Correlation heatmap
numeric_cols = ['Runs', 'Average', 'Strike_Rate', 'Centuries', 'Half_Centuries', 
               'Wickets', 'Bowling_Average', 'Economy']
correlation_data = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1)
plt.title('Performance Metrics Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 9. Key Insights

```python
print("="*60)
print("KEY INSIGHTS AND FINDINGS")
print("="*60)

# 1. Run difference in wins vs losses
won_avg_runs = won_matches['Runs'].mean()
lost_avg_runs = lost_matches['Runs'].mean()
print(f"\n1. Teams score {won_avg_runs - lost_avg_runs:.0f} more runs on average in winning matches")
print(f"   - Average in Won: {won_avg_runs:.2f}")
print(f"   - Average in Lost: {lost_avg_runs:.2f}")

# 2. Strike rate importance
won_sr = won_matches['Strike_Rate'].mean()
lost_sr = lost_matches['Strike_Rate'].mean()
print(f"\n2. Strike Rate is higher by {won_sr - lost_sr:.2f} in winning matches")
print(f"   - Average in Won: {won_sr:.2f}")
print(f"   - Average in Lost: {lost_sr:.2f}")

# 3. Best batting team
best_batting_team = team_stats.index[0]
print(f"\n3. Best Batting Team: {best_batting_team} (Total: {team_stats.iloc[0]:.0f} runs)")

# 4. Role distribution
print(f"\n4. Player Role Distribution:")
for role, count in df['Role'].value_counts().items():
    print(f"   - {role}: {count} ({count/len(df)*100:.1f}%)")

print("\n" + "="*60)
```

## 10. Conclusion

This analysis has revealed several key factors that contribute to match-winning performance:

1. **Runs Matter**: Teams scoring higher runs have better chances of winning
2. **Strike Rate is Critical**: Aggressive batting (higher strike rate) correlates with match wins
3. **Bowling Economy**: Lower economy rates in bowling lead to better match outcomes
4. **All-Rounders**: Players who contribute in both batting and bowling are valuable assets
5. **Consistency**: Players with higher averages show more consistent performance

These insights can help teams:
- Build balanced squads with strong batsmen and bowlers
- Focus on aggressive batting strategies
- Invest in all-rounders for flexibility
- Develop players who can perform consistently under pressure
